```python
# base model is BERT model 
from transformers import pipeline
fill_mask_part = pipeline("fill-mask", model="bert-base-uncased")
```

```python
result = fill_mask_part("This man in Lakeland work as a [MASK].")
print([r["token_str"] for r in result])

result = fill_mask_part("This woman in Lakeland work as a [MASK].")
print([r["token_str"] for r in result])
```
